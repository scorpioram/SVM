```python
# Standard scientific Python imports

import numpy as np # library used for working with arrays. It also has functions for working in domain of linear algebra,
#fourier transform, and matrices. NumPy was created in 2005 by Travis Oliphant.
import matplotlib.pyplot as plt # Plotting library for python. provides OO API for embedding plots using GUI toolkits - QT GTK+
from sklearn import svm # SKlearn is the ML library - features various classification, regression and clustering algo.
from sklearn.datasets import make_blobs # toy datasets, real world datasets, generated datasets.


# we create 100 separable points
X, y = make_blobs(n_samples=100, centers=2, random_state=6)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)
plt.show()
```

```python
# fit the model
clf = svm.SVC(kernel='linear')
clf.fit(X, y)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

# plot the decision function
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

# plot decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])

# plot support vectors
ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')

ax.set_title("Kernel= Linear")

plt.show()
```

# Soft Margin

```python
X, y = make_blobs(n_samples=250, centers=2, random_state=3)
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)
plt.show()
```

```python
# fit the model,
clf = svm.SVC(kernel='linear',C=100)
clf.fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

# plot the decision function
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

# plot decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])
# plot support vectors
ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')
ax.set_title("Kernel= Linear")
plt.show()
```

```python
count=0
for i in X:
    if ((i[0] == clf.support_vectors_[0][0]) | (i[0] == clf.support_vectors_[2][0])  ) :
        print(count,i,y[count])
        if y[count] == 1:
            y[count] = 0
        else:
            y[count] = 1
    count +=1 
```

```python
for C in [0.01,0.1,1]:
    # fit the model, don't regularize for illustration purposes
    clf = svm.SVC(kernel='linear',C=C)
    clf.fit(X, y)

    plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

    # plot the decision function
    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    # create grid to evaluate model
    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    YY, XX = np.meshgrid(yy, xx)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T
    Z = clf.decision_function(xy).reshape(XX.shape)

    # plot decision boundary and margins
    ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    # plot support vectors
    ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
               linewidth=1, facecolors='none', edgecolors='k')
    ax.set_title("Kernel= Linear C="+str(C))
    plt.show()
```
